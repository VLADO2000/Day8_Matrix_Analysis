```python
def correlation(dataset, threshold):
    #Creating dataframe to store result
    col_corr = pd.DataFrame()
    #Correlational matrix
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            #If correlation are higher than number and columns haven't been written yet
            if(corr_matrix.iloc[i,j]>=threshold) and (corr_matrix.columns[j] not in col_corr):
                #getting the names of the columns
                colname1 = corr_matrix.columns[i]
                colname2 = corr_matrix.columns[j]
                #corralation value
                corr_val = corr_matrix.iloc[i,j]
                #temporary dataframe 
                temp = pd.DataFrame({"column1":colname1,
                                     "column2":colname2,
                                    "corr":corr_val}, index=[0])
                #adding results in general df, also drop indexs 
                col_corr = pd.concat([col_corr, temp], ignore_index=True)
    return col_corr
```
