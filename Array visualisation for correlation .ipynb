```python
#Function wich helps to find corralation
def correlation(dataset, threshold):
    #Creating dataframe to store result
    col_corr = pd.DataFrame()
    #Correlational matrix
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            #If corralation are higher than and columns haven't been written yet
            if(corr_matrix.iloc[i,j]>=threshold) and (corr_matrix.columns[j] not in col_corr):
                #getting the name of the column
                colname = corr_matrix.columns[i] +'-' + corr_matrix.columns[j]
                #corralation value
                corr_val = corr_matrix.iloc[i,j]
                #temporary dataframe 
                temp = pd.DataFrame({"columns":colname,
                                    "corr":corr_val}, index=[0])
                #adding results in general df, also drop indexs 
                col_corr = pd.concat([col_corr, temp], ignore_index=True)
    return col_corr
```

```python
#col should be a simple string
#w_search represents name of the column in dataframe where search have to be done
def zoomming(df_copy, col, w_search):
    #Creating new column to store boolen output about exist sush  column name in Series or not
    df_copy[col] = df_copy.loc[:,w_search].str.contains(col)
    #Shrinking datasate to smaller one only with searched column name in there 
    df_copy=df_copy[df_copy[col]==True]
    df_copy.reset_index(drop=True, inplace=True)
    return df_copy
```
